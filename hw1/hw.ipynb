```python
import json
import csv

purchases = {}

with open('purchase_log.txt','r',encoding = 'utf-8') as f:
    f.readline()
    for line in f:
        line = line.strip()
        dictionary = json.loads(line)
        purchases[dictionary['user_id']] = dictionary['category']

with open('visit_log.csv','r',encoding = 'utf-8') as f:
    with open('funnel.csv','w',encoding = 'utf-8', newline='') as f2write:
        reader = csv.reader(f)
        writer = csv.writer(f2write)
        
        headers = next(reader)
        headers.append('category')
        writer.writerow(headers)
        
        for row in reader:
            user_id = row[0]
            if user_id in purchases:
                new_row = row + [purchases[user_id]]
                writer.writerow(new_row)
```

```python
count = 0
for user_id, category in purchases.items():
    print(f"{user_id} '{category}'")
    count += 1
    if count == 2:
        break
```
